```python
def aumentopeso(val_energia, masa, indice):
    energia = (masa / indice) * 300
    aumento = energia / (0.4 * val_energia)
    return aumento

def main():
    print("Bienvenido al sistema de nutricion Animal")

    while True:
        print("Ingrese el tipo de animal según las opciones:")
        print("1 - Vaca")
        print("2 - Cerdo")
        print("3 - Pollo")
        print("Ingrese 0 para salir del programa")
        opcion1 = int(input())

        if opcion1 < 0 or opcion1 > 3:
            print("Error: la opción ingresada es inválida")
        else:
            print("La opción ingresada es válida")

            if opcion1 == 0:
                break

            if opcion1 == 1:
                print("Ingresaste la opción de vaca")
                print("Ingrese tipo de alimento:")
                print("1 - Concentrado, 2 - Pasto")
                opcion2 = int(input())

                if opcion2 == 1:
                    print("Ingresaste opción concentrado")
                    print("Ingrese la cantidad de concentrado en kg:")
                    masa = float(input())
                    val_ener = 300
                    indice = 0.1
                elif opcion2 == 2:
                    print("Ingresaste opción pasto")
                    print("Ingrese la cantidad de pasto en kg:")
                    masa = float(input())
                    val_ener = 150
                    indice = 0.1
                else:
                    print("Error: la opción ingresada no existe")
                    continue

            elif opcion1 == 2:
                print("Ingresaste la opción de cerdo")
                print("Ingrese tipo de alimento:")
                print("1 - Concentrado, 2 - Cuido engorde")
                opcion2 = int(input())

                if opcion2 == 1:
                    print("Ingresaste opción concentrado")
                    print("Ingrese la cantidad de concentrado en kg:")
                    masa = float(input())
                    val_ener = 150
                    indice = 0.9
                elif opcion2 == 2:
                    print("Ingresaste opción cuido engorde")
                    print("Ingrese la cantidad de cuido engorde en kg:")
                    masa = float(input())
                    val_ener = 150
                    indice = 0.1
                else:
                    print("Error: la opción ingresada no existe")
                    continue

            elif opcion1 == 3:
                print("Ingresaste la opción de pollo")
                print("Ingrese tipo de alimento:")
                print("1 - Cuido levante, 2 - Cuido engorde")
                opcion2 = int(input())

                if opcion2 == 1:
                    print("Ingresaste opción cuido levante")
                    print("Ingrese la cantidad de cuido levante en gr:")
                    masa = float(input())
                    val_ener = 100
                    indice = 0.08
                elif opcion2 == 2:
                    print("Ingresaste opción cuido engorde")
                    print("Ingrese la cantidad de cuido engorde en gr:")
                    masa = float(input())
                    val_ener = 100
                    indice = 0.08
                else:
                    print("Error: la opción ingresada no existe")
                    continue

            aumento_peso = aumentopeso(val_ener, masa, indice)
            print("La cantidad de peso aumentado por día en gramos es:", aumento_peso)

if __name__ == "__main__":
    main()
```
